```python
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
n = 500
```

```python
# Set start and end date for the random date window
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2024-12-31')

# Generate all dates in that range
all_dates = pd.date_range(start=start_date, end=end_date)

# Randomly select 'n' dates from that range
random_dates = np.random.choice(all_dates, size=n)
```

```python
# Create the data dictionary
data = {
    'Date': random_dates,
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}

# Create the DataFrame
df = pd.DataFrame(data)

# Optional: Sort by date to make it look chronological
df.sort_values(by='Date', inplace=True)

# Preview the result
df.head()
```

```python
import pandas as pd

# Assuming 'data' was created in Task 2
df = pd.DataFrame(data)

# Show the first 5 rows
print("ðŸ”¹ DataFrame Head:")
print(df.head())

# Show dataframe info (structure, column types, non-null counts)
print("\nðŸ”¹ DataFrame Info:")
df.info()

# Show descriptive statistics for numeric columns (e.g., Likes)
print("\nðŸ”¹ DataFrame Description:")
print(df.describe())

# Show count of each category
print("\nðŸ”¹ Category Counts:")
print(df['Category'].value_counts())
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Likes'] = df['Likes'].astype(int)
```

```python
# Optional: Confirm the changes
print("\nðŸ”¹ Data Types After Cleaning:")
print(df.dtypes)

print("\nðŸ”¹ Data Preview After Cleaning:")
print(df.head())
```

```python
sns.histplot(df['Likes'], bins=30, kde=True)  # You can adjust bins and enable KDE for smooth curve
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Optional: make it wider for readability
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes Distribution per Category')
plt.xticks(rotation=45)  # Rotate x labels for better viewing
plt.show()
```

```python
print("ðŸ”¹ Mean Likes Across All Posts:")
print(df['Likes'].mean())
```

```python
print("\nðŸ”¹ Mean Likes per Category:")
print(df.groupby('Category')['Likes'].mean())
```

```python
# Save histogram
plt.figure()
sns.histplot(df['Likes'], bins=30, kde=True)
plt.title('Distribution of Likes')
plt.savefig('likes_histogram.png')

# Save boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes by Category')
plt.xticks(rotation=45)
plt.savefig('category_boxplot.png')
```
