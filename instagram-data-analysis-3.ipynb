```python
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
n = 500
```

```python
# Set start and end date for the random date window
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2024-12-31')

# Generate all dates in that range
all_dates = pd.date_range(start=start_date, end=end_date)

# Randomly select 'n' dates from that range
random_dates = np.random.choice(all_dates, size=n)
```

```python
# Create the data dictionary
data = {
    'Date': random_dates,
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}

# Create the DataFrame
df = pd.DataFrame(data)

# Optional: Sort by date to make it look chronological
df.sort_values(by='Date', inplace=True)

# Preview the result
df.head()
```

```python
import pandas as pd

# Assuming 'data' was created in Task 2
df = pd.DataFrame(data)

# Show the first 5 rows
print("🔹 DataFrame Head:")
print(df.head())

# Show dataframe info (structure, column types, non-null counts)
print("\n🔹 DataFrame Info:")
df.info()

# Show descriptive statistics for numeric columns (e.g., Likes)
print("\n🔹 DataFrame Description:")
print(df.describe())

# Show count of each category
print("\n🔹 Category Counts:")
print(df['Category'].value_counts())
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Likes'] = df['Likes'].astype(int)
```

```python
# Optional: Confirm the changes
print("\n🔹 Data Types After Cleaning:")
print(df.dtypes)

print("\n🔹 Data Preview After Cleaning:")
print(df.head())
```

```python
sns.histplot(df['Likes'], bins=30, kde=True)  # You can adjust bins and enable KDE for smooth curve
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Optional: make it wider for readability
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes Distribution per Category')
plt.xticks(rotation=45)  # Rotate x labels for better viewing
plt.show()
```

```python
print("🔹 Mean Likes Across All Posts:")
print(df['Likes'].mean())
```

```python
print("\n🔹 Mean Likes per Category:")
print(df.groupby('Category')['Likes'].mean())
```

```python
# Save histogram
plt.figure()
sns.histplot(df['Likes'], bins=30, kde=True)
plt.title('Distribution of Likes')
plt.savefig('likes_histogram.png')

# Save boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes by Category')
plt.xticks(rotation=45)
plt.savefig('category_boxplot.png')
```

```python
category_engagement_ratio = df.groupby('Category').agg(
    total_likes=('Likes', 'sum'),
    num_posts=('Likes', 'count')
)
category_engagement_ratio['likes_per_post'] = category_engagement_ratio['total_likes'] / category_engagement_ratio['num_posts']
```

```python
print(category_engagement_ratio)
```

```python
plt.figure(figsize=(10, 6))
daily_likes = df.groupby(df['Date'].dt.date)['Likes'].sum()
daily_likes.plot(kind='line', color='orange', marker='o')
plt.title('Total Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Extract the hour and day of the week from the 'Date' column
df['hour'] = df['Date'].dt.hour
df['day_of_week'] = df['Date'].dt.day_name()

# Pivot the data for the heatmap
heatmap_data = df.pivot_table(values='Likes', index='day_of_week', columns='hour', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)
plt.title('Average Likes by Hour and Day of Week')
plt.show()
```

```python
category_counts = df['Category'].value_counts()
plt.figure(figsize=(8, 8))
category_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3', len(category_counts)))
plt.title('Distribution of Categories in the Dataset')
plt.ylabel('')
plt.show()
```

```python
sns.pairplot(df[['Likes', 'hour', 'day_of_week']])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Likes', data=df, estimator='mean', palette='muted')
plt.title('Average Likes by Category')
plt.xlabel('Category')
plt.ylabel('Average Likes')
plt.xticks(rotation=45)
plt.show()
```

```python
top_posts = df.nlargest(10, 'Likes')
plt.figure(figsize=(10, 6))
sns.barplot(x='Likes', y='Category', data=top_posts, palette='viridis')
plt.title('Top 10 Posts with the Most Likes')
plt.xlabel('Likes')
plt.ylabel('Category')
plt.show()
```

```python
corr = df[['Likes', 'hour']].corr()
plt.figure(figsize=(6, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Likes and Hour')
plt.show()
```

```python
category_time = df.groupby([df['Date'].dt.date, 'Category'])['Likes'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.lineplot(data=category_time, x='Date', y='Likes', hue='Category', marker='o')
plt.title('Engagement Trends Over Time by Category')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```
